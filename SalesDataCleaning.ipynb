## Sales Trend

```python
monthly_sales = df.groupby(df['order date'].dt.to_period('M'))['total revenue'].sum()
monthly_sales.plot(title='Monthly Sales Trend')
```

## Top Category

```python
top_category = df.groupby('category')['total revenue'].sum()
top_category.sort_values(ascending=False).plot(kind='bar')
plt.xlabel('Category')
plt.ylabel('Total Revenue')
```

## Sales by Region

```python
region_sales = df.groupby('region')['total revenue'].sum().sort_values(ascending=False)
region_sales.plot(kind='bar')
```

## Online vs offline Sales

```python
channel_sales = df.groupby('sales channel')['total revenue'].sum().sort_values(ascending=False)
channel_sales
```
